```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your combined CSV
df = pd.read_csv('agricultural_exports_combined.csv')

# Group by 'Year' and sum 'Trade Value'
agri_exports_per_year = df.groupby('Year')['Trade Value'].sum()

# Plot
plt.figure(figsize=(12, 6))
plt.plot(agri_exports_per_year.index, agri_exports_per_year.values, marker='o')
plt.title('Total Agricultural Exports per Year')
plt.xlabel('Year')
plt.ylabel('Trade Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.bar(agri_exports_per_year.index, agri_exports_per_year.values)
plt.title('Total Agricultural Exports per Year')
plt.xlabel('Year')
plt.ylabel('Trade Value')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
from scipy import signal

# Detrend the export values
detrended_values = signal.detrend(agri_exports_per_year.values)

# Plot the detrended data
plt.figure(figsize=(14, 7))
plt.plot(agri_exports_per_year.index, detrended_values, marker='o')
plt.title('Detrended Agricultural Exports per Year')
plt.xlabel('Year')
plt.ylabel('Detrended Trade Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Difference the series (year-on-year changes)
diff_values = agri_exports_per_year.diff().dropna()

# Plot differenced data
plt.figure(figsize=(14, 7))
plt.bar(diff_values.index, diff_values.values)
plt.title('Year-on-Year Change in Agricultural Exports')
plt.xlabel('Year')
plt.ylabel('Change in Trade Value')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
